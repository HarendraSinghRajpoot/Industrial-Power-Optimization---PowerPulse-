```python
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Example true labels and predicted labels
true_labels = np.array([0, 1, 2, 2, 0, 1, 0, 2, 1, 0])
predicted_labels = np.array([0, 2, 2, 2, 0, 0, 0, 2, 1, 1])

# Generate the confusion matrix
cm = confusion_matrix(true_labels, predicted_labels)

# Print the confusion matrix
print("Confusion Matrix:\n", cm)

# Initialize the plot
plt.figure(figsize=(8, 6))
ax = sns.heatmap(cm, annot=False, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1', 'Class 2'], yticklabels=['Class 0', 'Class 1', 'Class 2'])

# Calculate TP, FP, FN, and TN for each class and annotate the heatmap
classes = [0, 1, 2]
for i in range(len(classes)):
    for j in range(len(classes)):
        TP = cm[i, i]
        FP = cm[:, j].sum() - cm[j, j]
        FN = cm[i, :].sum() - cm[i, i]
        TN = cm.sum() - (TP + FP + FN)
        
        if i == j:
            label = f'TP={TP}'
        elif i != j and cm[i, j] != 0:
            label = f'FP={cm[i, j]}\nFN={cm[j, i]}'
        else:
            label = f'TN={TN}'
        
        ax.text(j + 0.5, i + 0.5, label, ha='center', va='center', color='black')

plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix for 3 Classes with Annotations')
plt.show()
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import f1_score

# Create a synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the models
model1 = RandomForestClassifier(random_state=42)
model2 = GradientBoostingClassifier(random_state=42)

# Train the models
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)

# Predict on the test set
y_pred1 = model1.predict(X_test)
y_pred2 = model2.predict(X_test)

# Calculate the F1 score for both models
f1_score_model1 = f1_score(y_test, y_pred1)
f1_score_model2 = f1_score(y_test, y_pred2)

# Print the F1 scores
print(f"F1 Score for RandomForestClassifier: {f1_score_model1}")
print(f"F1 Score for GradientBoostingClassifier: {f1_score_model2}")

# Determine which model is better
if f1_score_model1 > f1_score_model2:
    print("RandomForestClassifier is better.")
else:
    print("GradientBoostingClassifier is better.")
```
